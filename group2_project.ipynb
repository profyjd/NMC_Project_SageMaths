A ball has initial temperature at 1000K, what will be its temp after 500 secs?
The differential equation of the temp of the ball is given by "Differential equation:
dθ/dt = -2.2067e-12 * (θ⁴ - 81e8)"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, theta):
    "Differential equation: dθ/dt = -2.2067e-12 * (θ⁴ - 81e8)""
    return -2.2067e-12 * (theta**4 - 81e8)
```

```python
# Initial values
t0 = 0               # Initial time (s)
theta0 = 1000.0      # Initial temperature (K)
h = 250              # Step size (s)
steps = 2            # Number of steps to reach t = 500 seconds
```

```python
# For storing results
t = t0
theta = theta0

print("Step\tTime (s)\tTheta (K)")
print(f"0\t{t:.0f}\t\t{theta:.4f}")

for i in range(1, steps + 1):
    k1 = round(f(t, theta), 6)
    k2 = round(f(t + h / 2, theta + h * k1 / 2), 6)
    k3 = round(f(t + h / 2, theta + h * k2 / 2), 6)
    k4 = round(f(t + h, theta + h * k3), 6)

    theta += (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    t += h

    # Round θ to 4 decimal places as per instruction
    theta = round(theta, 4)

    print(f"{i}\t{t:.0f}\t\t{theta:.4f}")
```

```python
# prompt: plot graph for different values of time t

# We need to store the t and theta values to plot them
t_values = [t0]
theta_values = [theta0]

t = t0
theta = theta0

for i in range(1, steps + 1):
    k1 = round(f(t, theta), 6)
    k2 = round(f(t + h / 2, theta + h * k1 / 2), 6)
    k3 = round(f(t + h / 2, theta + h * k2 / 2), 6)
    k4 = round(f(t + h, theta + h * k3), 6)

    theta += (h / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    t += h

    # Round θ to 4 decimal places
    theta = round(theta, 4)

    t_values.append(t)
    theta_values.append(theta)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t_values, theta_values, marker='o', linestyle='-')
plt.title('Temperature (θ) vs. Time (t)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.grid(True)
plt.show()
```
